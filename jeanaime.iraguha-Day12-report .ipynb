### Exercise 7:
Plot two line plots on the same graph: `y1 = x^2` and `y2 = x^3` for `x` ranging from 0 to 10.

```python
# Exercise 7
```

---

````python
import numpy as np
X= np.linspace(0,10)
y1=X**2
y2=X**3
plt.plot(X, y1, color="blue", linewidth=2.5, linestyle="-", label="square")
plt.plot(X, y2, color="red",  linewidth=2.5, linestyle="-", label="cube")
plt.legend()
plt.title("cubic and Square function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
# Exercise 9
```

---

````python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values
y1 = x**2
y2 = x**3

# Create a figure and a set of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Plot y1 = x^2 in the first subplot
ax1.plot(x, y1, label='$y = x^2$', color='b')
ax1.set_title('Plot of $y = x^2$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Plot y2 = x^3 in the second subplot
ax2.plot(x, y2, label='$y = x^3$', color='r')
ax2.set_title('Plot of $y = x^3$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
````

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

```python
# Exercise 10
```

---

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values
y1 = x
y2 = x**2
y3 = x**3

# Create a figure
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='$y = x$', color='b', linestyle='-')

plt.plot(x, y2, label='$y = x^2$', color='g', linestyle='--')


plt.plot(x, y3, label='$y = x^3$', color='r', linestyle=':')

# Add title and labels
plt.title('Plots of $y = x$, $y = x^2$, and $y = x^3$')
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Show the plot
plt.show()
````

### Exercise 11:
Create a pie chart with 5 segments labeled A, B, C, D, and E, with values `[25, 35, 20, 15, 5]`.

```python
# Exercise 11
```

---

````python
import matplotlib.pyplot as plt

# Data to plot
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [25, 35, 20, 15, 5]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0, 0) 
# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=False)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart with 5 Segments')
plt.show()
````

### Exercise 12:
Create a horizontal bar chart with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`.

```python
# Exercise 12
```

---

````python
import matplotlib.pyplot as plt

# Data to plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color=['blue', 'green', 'red', 'purple'])

# Add title and labels
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')

# Show the plot
plt.show()
````
